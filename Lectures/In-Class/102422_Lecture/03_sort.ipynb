```python
# merge_sort

def merge(numbers, p, q, r):
    left = numbers[p:q+1]
    right = numbers[q+1:r+1]
    
    i = 0 # index into left
    j = 0 # index into right
    k = p # index into numbers[p:r+1]
    
    while i < len(left) and j < len(right):
        if left[i] < right[j]:
            numbers[k] = left[i]
            i = i + 1
        else:
            numbers[k] = right[j]
            j = j + 1
        k = k + 1
        
    if i < len(left):
        numbers[k:r+1] = left[i:]
        
    if j < len(right):
        numbers[k:r+1] = right[j:]
        
    print("MERGE  : " + str(numbers)) # print result of each merge

def merge_sort(numbers, p = None, r = None):
    if p == None and r == None:
        p = 0
        r = len(numbers) - 1
        
    if p < r:
        q = (p + r) // 2             # Note: lecture video is missing a '/'
        print("PART   : " + str(numbers[p:q+1]))
        merge_sort(numbers, p, q)
        print("PART   : " + str(numbers[q+1:r+1]))
        merge_sort(numbers, q+1, r)
        merge(numbers, p, q, r)

numbers = [10, 4, 78, 2, 45, 32, 7, 11]
print("START  : " + str(numbers))
merge_sort(numbers)
print("FINISH : " + str(numbers))

```

```python
# Quicksort

# Partition the sublist numbers[p ... r] so that the pivot
# (originally in numbers[r]) moves to numbers[q],
# all items in numbers[p ... q-1] are less than or equal to the pivot,
# and all items in numbers[q+1 ... r] are greater than the pivot.
# Return the index q where the pivot ends up.

def partition(numbers, p, r):
    
    # In this implementation, always pick last item in list as the pivot.
    # Question: what are other options for choosing the pivot?
    # What are the considerations for the different options?
    
    pivot = numbers[r]
    print("Pivot  : " + str(pivot))

    
    # Set up the indices i and j so that
    #    numbers[p ... i] contains items <= pivot,
    #    numbers[i+1 ... j-1] contains items > pivot, and
    #    numbers[j ... r-1] contains items not yet compared with the pivot.

    i = p - 1
    j = p

    while j < r:
        if (numbers[j] < pivot):
            # Move this item into the section known to be <= pivot.
            i += 1
            (numbers[i], numbers[j]) = (numbers[j], numbers[i])
        j += 1
            
    # Get the pivot into the correct position.
    (numbers[i+1], numbers[r]) = (numbers[r], numbers[i+1])
    print("ROUND  : " + str(numbers))
    return i+1

# Sort the sublist numbers[p ... r] using the quicksort algorithm.
def quick_sort(numbers, p = None, r = None):
    if p == None and r == None:
        p = 0
        r = len(numbers) - 1
    
    if p < r:   # nothing to do if the sublist has fewer than 2 items
        q = partition(numbers, p, r) # divide
        quick_sort(numbers, p, q-1)   # conquer smaller items
        quick_sort(numbers, q+1, r)   # conquer larger items


numbers = [10, 4, 78, 2, 45, 32, 7, 11]
print("START  : " + str(numbers))
quick_sort(numbers)
print("FINISH : " + str(numbers))
```
