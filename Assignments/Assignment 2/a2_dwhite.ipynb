## Q1. (4 points)
 Bubble sort is a simple sorting algorithm that works by repeatedly stepping through a list to be sorted, comparing each pair of adjacent items and swapping them if they are in the wrong order. This process is repeated until after a full pass through the list where no items are swapped. Below is an example of how bubble sort works.

First Pass

`( 5 1 4 2 8 ) → ( 1 5 4 2 8 )`, compare the first two elements, and swap since 5 > 1

`( 1 5 4 2 8 ) → ( 1 4 5 2 8 )`, swap since 5 > 4

`( 1 4 5 2 8 ) → ( 1 4 2 5 8 )`, swap since 5 > 2

`( 1 4 2 5 8 ) → ( 1 4 2 5 8 )`, do not swap since 5 < 8

 

Second Pass

`( 1 4 2 5 8 ) → ( 1 4 2 5 8 )`

`( 1 4 2 5 8 ) → ( 1 2 4 5 8 )`, swap since 4 > 2

`( 1 2 4 5 8 ) → ( 1 2 4 5 8 )`

`( 1 2 4 5 8 ) → ( 1 2 4 5 8 )`, we swapped on this iteration, so one more iteration is needed

 

Third Pass

`( 1 2 4 5 8 ) → ( 1 2 4 5 8 )`

`( 1 2 4 5 8 ) → ( 1 2 4 5 8 )`

`( 1 2 4 5 8 ) → ( 1 2 4 5 8 )`

`( 1 2 4 5 8 ) → ( 1 2 4 5 8 )`, no swaps on this iteration, so we are done

 


##### (a) Write a Python function, `bubble_sort`, that performs bubble sort. Verify that your function works by testing it on the following four lists, printing the lists before and after the sort:

- `list1 = [5,1,4,2,8]`
- `list2 = ["aa","a","a1","bat","baa","ba","b"]`
- `list3 = ["5","1","11","3","03"]`
- a list of twenty randomly generated integers between 0 and 100.


```python
## RUN TIME COMPLEXITY OF BUBBLE SORT ##
# It's O(n^2)
# Each loop of the program has a run time that is a function of the length of the list. Since our list can be infinite, and we have a loop with a nested loop, it's n * n = n^2.
# while True worst case is n, and the for loops worst case is n. 

# import random for the list 4
import random 

## need to define bubble_sort
def bubble_sort(sortList): 
    
    # using an infinite because we'll break the loop later
    while True:
        # make the flag to indicate we haven't done swaps
        no_swaps = True
        # loop over the list
        for i in range(0, len(sortList) - 1):
            # check to see if they need to be swap
            if sortList[i] > sortList[i + 1]:
                # swap them 
                sortList[i], sortList[i + 1] = sortList[i + 1], sortList[i]
                # set our flag for this iteration, because we've already done at least one swap
                no_swaps = False
        # break as soon as we go through an iteration that hasn't swapped
        if no_swaps:
            break
        # print it for each step 
        print(sortList)

list1 = [5,1,4,2,8]
list2 = ["aa","a","a1","bat","baa","ba","b"]
list3 = ["5","1","11","3","03"]

list4 = []
for i in range(20):
    list4.append(random.randrange(0, 101))


# added the prints at the end of each to verify the sorting outcome because I don't know how 
# alphanumerics or ints as strings are ordered :) 
# feel free to uncomment if you want to verify any 

bubble_sort(list1)
#print(sorted(list1))
bubble_sort(list2)
#print(sorted(list2))
bubble_sort(list3)
#print(sorted(list3))
bubble_sort(list4)
#print(sorted(list4))

```

Implement a recursive binary search function, `binary_search`, which allows you to search a sorted list of items for a specific target.

Your recursive function should take four input parameters: `list` (a sorted list), `key` (the item being searched for), a `left_index` into the list, and a `right_index` into the list. Your function looks only in the sublist starting at the `left_index`, and ending at the `right_index` (including the item at the `right_index` position). 

For example, suppose that the_list is:

`[27, 78, 105, 135, 411, 431, 434, 468, 493, 501, 525, 534, 551, 654, 780]`

which contains 15 items, indexed 0 to 14.



 If I call `binary_search(the_list, 135, 0, 14)`, then I'm considering the entire list, and the return value should be 3, since the key `135` appears at index 3.
 
  If I call `binary_search(the_list, 500, 0, 14)`, the return value should be `None`, since the key `500` is not in `the_list`. The call `binary_search(the_list, 135, 2, 12)` should return 3 since, as before, the key `135` appears at index 3 in `the_list`. 
  
  Notice that the index returned is the index in the entire list, not the relative index in the sublist starting at `left_index`. The parameters `left_index` and `right_index` just give the range of indices into the list that we are considering in a given call of `binary_search`.


##### Here is what you need to do in your `binary_search` function:

- If the sublist of `the_list` starting at `left_index` and ending at `right_index` is empty, then the key cannot be in the sublist -- return `None`.
- Otherwise, the sublist is not empty, and you should:
    - Compute the midpoint of this sublist, by taking the average of `left_index` and `right_index`, and rounding to the nearest integer as needed.
    - If `key` is equal in value to the item at midpoint, then it has been found. Return this index.
    - Otherwise, because `the_list` is sorted, you can tell whether `key`, if in `the_list`, is either in the sublist before `midpoint` or in the sublist after `midpoint`.
    - If key is less than the item at midpoint and it's in the list, then it must be in the sublist before  `midpoint`. Recursively return the result of calling `binary_search` on the sublist starting at `left_index` and going up to and including the index just before `midpoint`.
    - Only one other possibility remains: `key` is greater than the item at `midpoint`, and so if it's in the list, then it must be in the sublist after `midpoint`. Recursively return the result of calling `binary_search` on the sublist starting at the index just after `midpoint` and going up to and including `right_index`.

##### Once you have written your recursive binary search function, you should write a simple user interface as follows:

- Prompt the user to specify the number of elements `(N)` in a list that you will create to be searched.
- Generate a list of `N` random integers, with each integer in the range of 0 to 1000, then sort it. (You can use your very own `bubble_sort` function from Q1 above, or Python's built-in function sort() or sorted, to sort this list).
- Print the list so that the user can see it.
- Prompt the user for input from the keyboard. If the user enters "?", print the list. If the user enters "q", quit the search interface. Otherwise search the list for the element that the user entered.
- Repeat step 4 until the user quits.

```python
import random

# define the function
def binary_search(the_list, key, left_index, right_index):
    # doing > because the left_index shouldn't be greater than right index anyway, if they're the same
    # there's one element because indexing 
    if left_index > right_index:
        return None
   
    # printing for testin
    # print(the_list)
    # calculate the midpoint, round down
    midpoint = (left_index + right_index) // 2
    # print("The midpoint is:", midpoint)
    # printing for checking the math, looks good 
    #print(midpoint)

    # check if our midpoint is the key
    if the_list[midpoint] == key:
        return midpoint
    # if midpoint is greater than our target, then recur on the left half of the list
    elif the_list[midpoint] > key:
        return binary_search(the_list, key, left_index, midpoint - 1)
    # if midpoints is less than our target, then recur on the left half of the list
    elif the_list[midpoint] < key:
        return binary_search(the_list, key, midpoint + 1, right_index)
    # otherwise, it's not there
    else:
        return None

# now we make the input for taking the function
def binarySearchInput():
    # define list
    out_list = []
    # ask for n
    num = int(input("How many elements would you like me to add to our list?"))
    # make n of those between 1, 1000
    for i in range(num):
        # append them
        out_list.append(random.randint(0, 1000))
    # sourt it
    out_list.sort()
    # print the final list
    print("Your list is:", out_list)
    # keep it running
    while True:
        # take input for key
        in_key = input("What would you like to search for? (enter q for quit or ? to print the list again")
        
        if in_key == "q":
            break
        elif in_key == "?":
            print(out_list)
        else: 
            # run binary_search on their key!
            print(binary_search(out_list, int(in_key), 0, len(out_list)))

## MAIN 
the_list = [27, 78, 105, 135, 411, 431, 434, 468, 493, 501, 525, 534, 551, 654, 780]
print(binary_search(the_list, 135, 0, 14)) # fine
print(binary_search(the_list, 500, 0, 14)) # fine 
print(binary_search(the_list, 135, 2, 12))
binarySearchInput()
```
